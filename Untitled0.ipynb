# Data Analysis

# Distribution of House Prices




```python
# Plot 1: Distribution of house prices
plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], bins=50, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

**Summary:**

*   This histogram illustrates the distribution of median house values across the dataset. The x-axis represents the median house value, and the y-axis shows the frequency of occurrences. The plot reveals a right-skewed distribution, with most houses priced below $300,000 and a significant spike at the maximum value, likely due to data capping. This visualization helps identify the general pricing trend and the presence of outliers or capped values.

# House Prices by Ocean Proximity

```python
# Plot 2: House Prices in Location
plt.figure(figsize=(10, 6))
sns.boxplot(x='ocean_proximity', y='median_house_value', data=df)
plt.title('House Prices by Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Median House Value')
plt.show()
```

**Summary**:
*   This boxplot illustrates the distribution of median house values based on their proximity to the ocean. The categories on the x-axis represent different levels of ocean proximity, while the y-axis shows the median house value. Houses located on islands tend to have the highest median values, followed by those near the ocean and near the bay. Inland properties have the lowest median values. The plot reveals that proximity to the ocean generally corresponds with higher house prices. Additionally, the presence of outliers, particularly in the inland category, indicates variability in house values within that group. This visualization highlights the premium associated with ocean proximity and the variability in housing prices across different regions.


# Correlation Heatmap

```python
# Plot 3: Correlation heatmap

numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Summary**:


* The heatmap displays the correlation coefficients between different numerical features in the dataset. Each cell represents the correlation between two variables, with values ranging from -1 to 1. Warmer colors indicate positive correlations, while cooler colors indicate negative correlations. Notably, there is a strong positive correlation between median_income and median_house_value (0.69), indicating that higher income levels are associated with higher house prices. Additionally, total_rooms and total_bedrooms show a high correlation (0.93), suggesting these features are closely related. The negative correlation between longitude and latitude (-0.92) reflects geographical trends across California.


# Median Income vs Median House Value

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='median_income', y='median_house_value', data=df)
plt.title('Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

**Summary**
*    This scatter plot illustrates the relationship between median income and median house value. The x-axis represents median income, while the y-axis shows the median house value. The plot reveals a positive correlation, indicating that higher income levels are generally associated with higher house prices. The clustering of points along the top suggests a capping effect on house prices at $500,000, which might limit the observed relationship at higher values. This plot highlights the significant influence of income on housing affordability and market value.

# Average House Price Over Time

```python
current_year = 2024
df['year_built'] = current_year - df['housing_median_age']
yearly_prices = df.groupby('year_built')['median_house_value'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(yearly_prices['year_built'], yearly_prices['median_house_value'], marker='o')
plt.title('Average House Price Over Time')
plt.xlabel('Year Built')
plt.ylabel('Average Median House Value')
plt.grid(True)
plt.show()
```

**Summary**
*   This line plot illustrates the trend in average median house values over time, inferred from the housing median age. The x-axis represents the estimated year of construction, while the y-axis shows the average median house value. The plot reveals fluctuations in house prices, with a noticeable decline in the late 1970s, relative stability through the 1980s and 1990s, and a significant rise in recent years. The sharp drop at the end may indicate data limitations or edge effects from the estimation process. Overall, the plot suggests that newer houses tend to have higher values, reflecting trends in the housing market over time.
